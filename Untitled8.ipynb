```python
import pandas as pd
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")
df.head()
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.columns
```

```python
df.columns
```

```python
import plotly.express as px

sales_country = df.groupby("COUNTRY")["SALES"].sum().reset_index()

fig = px.bar(sales_country, x="COUNTRY", y="SALES",
             title="Sales by Country")
fig.show()
```

```python
sales_product = df.groupby("PRODUCTLINE")["SALES"].sum().reset_index()

fig = px.bar(sales_product, x="PRODUCTLINE", y="SALES",
             title="Sales by Product Line")
fig.show()
```

```python
sales_year = df.groupby("YEAR_ID")["SALES"].sum().reset_index()

fig = px.line(sales_year, x="YEAR_ID", y="SALES",
              title="Sales Trend by Year")
fig.show()
```

```python
fig = px.pie(sales_country, names="COUNTRY", values="SALES",
             title="Country Contribution to Sales")
fig.show()
```
